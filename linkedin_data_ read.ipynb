# Data collection for Warsaw Shool of Economics

The purpose of the program is to download data on Warsaw School of Economics alumni employment after graduation. The data will be downloaded from 2010 to 2021. It can be assumed that in 2010 a significant number of students already had a LinkedIn account. Tidy data series that will be created in the future can be used to visualised stored data.
On the website one can find information such as: 
* place of residence,
* company where the graduate started working,
* type of position alumni,
* field of study alumni have completed,
* skills alumni declared,
* type of connections.

Data stored on university page shows only 15 most popular values. 
In tidy data place of residence and degree of connection have been omitted due to insignificans and difficulty in reading.

### Company function

```python
def alumni_company(b_year,e_year):
    c_fr=[]
    for year in range(b_year,(e_year)+1):
        #Empty boxes
        driver.find_element_by_id('people-search-year-start').clear()
        time.sleep(2)
        driver.find_element_by_id('people-search-year-end').clear()
        
        driver.find_element_by_id('people-search-year-start').send_keys(year)
        time.sleep(2)
        driver.find_element_by_id('people-search-year-end').send_keys(year)
        time.sleep(2)
        items=driver.find_elements_by_css_selector('div.insight-container')
        content=[]
        for item in items:
            text = item.text
            content.append(text)
        company1=content[1]
        company1_s=company1.splitlines()
        company1_s=company1_s[-15:]
        num=[]
        comp=[]
        yr=[]
        for i in range(0,(len(company1_s))):
            dat=company1_s[i].split(" ", 1)
            num.append(dat[0])
            comp.append(dat[1])
            yr.append(year)
        df = pd.DataFrame({'Alumni':num, 'Variable':comp, 'Year':yr, 'Description':'Company'})
        #print(df)
        c_fr.append(df)
    result = pd.concat(c_fr)
    return result
```
